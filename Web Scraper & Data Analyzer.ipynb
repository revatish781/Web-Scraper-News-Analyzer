```python

import pandas as pd
import random
import matplotlib.pyplot as plt
from collections import Counter


categories = ["Tech", "Business", "Sports", "Health", "Entertainment"]
keywords = ["AI", "Market", "Election", "Vaccine", "Movie", "Stock", "Game", "Startup", "Health", "Music"]

headlines = []
for i in range(200):
    category = random.choice(categories)
    title = " ".join(random.choices(keywords, k=random.randint(3,6)))
    headlines.append([category, title])


df = pd.DataFrame(headlines, columns=["Category","Headline"])
df.to_csv("synthetic_news.csv", index=False)
print("Synthetic news dataset saved as 'synthetic_news.csv'")


all_words = " ".join(df['Headline']).split()
word_counts = Counter(all_words)
top_words = word_counts.most_common(10)
print("Top 10 words in headlines:", top_words)


words, counts = zip(*top_words)
plt.figure(figsize=(8,5))
plt.bar(words, counts, color='skyblue')
plt.title("Top 10 Words in Headlines")
plt.ylabel("Frequency")
plt.show()


category_counts = df['Category'].value_counts()
plt.figure(figsize=(6,4))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Headline Distribution by Category")
plt.ylabel("")
plt.show()
```
